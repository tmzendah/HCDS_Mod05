# 03 — Quality Control

Runs image quality checks and flags poor-quality radiographs.

- Blank images (std < 5)
- Clipped / over-exposed images
- Conversion errors

Outputs: `reports/qc_reports/qc_report.csv`, `reports/flagged_images/`

```python
import subprocess
subprocess.run([
    'python', 'scripts/quality_control.py',
    '--image_dir', 'data/processed',
    '--output', 'reports/qc_reports/qc_report.csv'
])
subprocess.run([
    'python', 'scripts/flag_poor_quality.py',
    '--qc_report', 'reports/qc_reports/qc_report.csv',
    '--flag_dir', 'reports/flagged_images'
])
```
